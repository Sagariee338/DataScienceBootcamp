```python
# OOPS CONCEPTS
# Class is basically a blue print which has features and attributes
class Person:
  def __init__(self,a,b):
    self.a=a
    self.b=b
    print("Class Initialized")
  def sum(self):
    return self.a+self.b

person_object=Person(17,20)
print("==================")
print(person_object.sum())
print(person_object.a)
print(person_object.b)
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
training_X=[[7, 2, 3], [4, np.nan, 6], [10, 5, 9]]
imp_mean.fit(training_X)
X_test = [[np.nan, 2, 3], [4, np.nan, np.nan], [10, np.nan, 9]]
print(imp_mean.transform(X_test))
```
